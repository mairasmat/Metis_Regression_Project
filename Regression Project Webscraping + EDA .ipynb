```python
def get_movie_value(soup, field_name):
    
    '''Grab a value from Box Office Mojo HTML
    
    Takes a string attribute of a movie on the page and returns the string in
    the next sibling object (the value for that attribute) or None if nothing is found.
    '''
    
    obj = soup.find(text=re.compile(field_name))
    
    if not obj: 
        return None
    
    # this works for most of the values
    next_element = obj.findNext()
    
    if next_element:
        return next_element.text 
    else:
        return None
```

```python
import dateutil.parser

def money_to_int(moneystring): #gets revenue
    moneystring = moneystring.replace('$', '').replace(',', '')
    return int(moneystring)

def dist_cleaner(dist_name): #gets distributor name
    try:
        distributor = dist_name.split(' ')
        del distributor[:-4:-1]
        distributor = ' '.join(distributor)
        distributor = distributor.strip('See')
        return distributor
    except:
        return None

def runtime_to_minutes(runtimestring): #gets runtime
    runtime = runtimestring.split()
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None

def to_date(datestring): #gets date
    date = dateutil.parser.parse(datestring)
    return date
```

```python
def get_movie_dict(link):
    '''
    From BoxOfficeMojo link stub, request movie html, parse with BeautifulSoup, and
    collect 
        - title 
        - distributor
        - domestic gross
        - runtime 
        - MPAA rating
        - full release date
    # removed    - genre
    Return information as a dictionary.
    '''
    
    base_url = 'https://www.boxofficemojo.com'
    
    #Create full url to scrape
    url = base_url + link
    
    #Request HTML and parse
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page,"lxml")

    
    headers = ['Title', 'Distributor',
               'Runtime', 'Rating', 'Release_Date']
    
    #Get title
    title_string = soup.find('title').text
    title = title_string.split('-')[0].strip()

    #Get distributor
    distributor_string = get_movie_value(soup, 'Distributor')
    distributor = dist_cleaner(distributor_string)
    
    
    #Get runtime
    raw_runtime = get_movie_value(soup,'Running')
    if raw_runtime: 
        runtime = runtime_to_minutes(raw_runtime)
    else:
        runtime = np.nan
    
    #Get rating
    rating = get_movie_value(soup,'MPAA')

    #Get release date - decided to scrap due to late-stageparsing issues 
    #raw_release_date = get_movie_value(soup,'Release Date')
    #if raw_release_date:
     #   release_date = to_date(raw_release_date.split('\n')[0])  
    #else:
     #   release_date = np.nan
        
    #Create movie dictionary and return
    movie_dict = dict(zip(headers, [title,
                                distributor,
                                runtime,
                                rating]))

    return movie_dict
```

```python
gset=['Action','Adventure','Biography','Comedy','Crime','Documentary','Drama','Family','History','Horror',
      'Music','Mystery','Romance','Sport','Thriller','War','Western']
Action,Adventure,Biography,Comedy,Crime,Documentary,Drama,Family,History,Horror,Music,Mystery,Romance,Sport,Thriller,War,Western=([] for i in range(len(gset)))

glist=[Action,Adventure,Biography,Comedy,Crime,Documentary,Drama,Family,History,Horror,Music,Mystery,Romance,Sport,Thriller,War,Western]
for g in genres:
    for c in gset:
        if c in g:
            glist[gset.index(c)].append(1)
        else:
            glist[gset.index(c)].append(0)
```
